```python
import matplotlib.pyplot as plt

# Columns you want to plot
cols = ["precipitation", "temp_max", "temp_min", "wind"]

plt.figure(figsize=(12, 8))

for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, i)  # 2x2 grid
    plt.hist(df[col], bins=30, edgecolor="black", alpha=0.7)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your dataframe
cols = ["precipitation", "temp_max", "temp_min", "wind"]

plt.figure(figsize=(12, 8))

for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, i)  # 2x2 grid
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
df["precipitation_log"] = np.log1p(df["precipitation"])  
print(df["precipitation_log"])
```

```python
import matplotlib.pyplot as plt

# Compare precipitation before and after log transform
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(df["precipitation"], bins=30, edgecolor="black")
plt.title("Original precipitation")

plt.subplot(1,2,2)
plt.hist(df["precipitation_log"], bins=30, edgecolor="black")
plt.title("Log-transformed precipitation")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Create a small dataset
data = {
    "precipitation": [0, 0, 5, 10, 0, 2, 0, 20, 0, 15, 0, 3, 7, 0, 25],
    "temp_max": [12, 15, 20, 18, 25, 30, 28, 10, 16, 22, 24, 19, 14, 27, 21],
    "temp_min": [5, 7, 10, 9, 12, 15, 14, 3, 6, 11, 13, 8, 4, 12, 9],
    "wind": [2, 3, 4, 2.5, 5, 6, 4.5, 3.5, 2.8, 5.5, 6.2, 3.2, 4.8, 2.9, 5.1],
}

df = pd.DataFrame(data)

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df["precipitation"], bins=10, edgecolor="black", alpha=0.7)
plt.title("Distribution of Precipitation")
plt.xlabel("Precipitation (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df["wind"], bins=10, edgecolor="black", alpha=0.7)
plt.title("Distribution of wind")
plt.xlabel("wind ")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Two datasets
x1 = df["temp_max"]
x2 = df["temp_min"]

plt.figure(figsize=(10,6))

# Plot both histograms with transparency
plt.hist(x1, bins=10, alpha=0.6, label="Temp Max", edgecolor="black")
plt.hist(x2, bins=10, alpha=0.6, label="Temp Min", edgecolor="black")

plt.title("Grouped Histogram: Temp Max vs Temp Min")
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
