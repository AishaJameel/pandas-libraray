```python
import pandas as pd
import numpy as np

# Generate skewed data
np.random.seed(0)

data = {
    "Normal_Distribution": np.random.normal(50, 10, 100),   # Symmetric (mean ~ median)
    "Right_Skewed": np.random.exponential(10, 100),         # Long tail to the right
    "Left_Skewed": -np.random.exponential(10, 100) + 50     # Long tail to the left
}

df = pd.DataFrame(data)
print(df.head())

# Calculate skewness
print("\nSkewness of each column:")
print(df.skew())
```

```python
import matplotlib.pyplot as plt

# Histogram for each distribution
df.hist(bins=30, figsize=(12, 6))
plt.suptitle("Histograms Showing Skewness", fontsize=16)
plt.show()

# Boxplots for each distribution
plt.figure(figsize=(10, 6))
df.boxplot()
plt.title("Boxplots Showing Skewness", fontsize=16)
plt.ylabel("Values")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# One encoder for authors
le_author = LabelEncoder()
data1["author_encoded"] = le_author.fit_transform(data1["author"])

# Another encoder for narrators
le_narrator = LabelEncoder()
data1["narrator_encoded"] = le_narrator.fit_transform(data1["narrator"])

print(data1[["author", "author_encoded", "narrator", "narrator_encoded"]].head())
```

```python
# Example cleanup
data1["stars_clean"] = data1["stars"].str.extract(r'(\d+\.?\d*)').astype(float)
print(data1[["stars", "stars_clean"]])
```

```python
import re

def convert_time(t):
    hrs = re.search(r'(\d+)\s*hrs?', t)
    mins = re.search(r'(\d+)\s*mins?', t)
    total = 0
    if hrs:
        total += int(hrs.group(1)) * 60
    if mins:
        total += int(mins.group(1))
    return total

data1["time_mins"] = data1["time"].apply(convert_time)
print(data1["time_mins"],"Time in mins")
```
