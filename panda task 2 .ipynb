```python

#Can you split name into parts (e.g., series name vs. book numbers, or title vs subtitle) where appropriate?
import re

def split_name(name):
    title, subtitle, series, book_number = name, None, None, None
    
    # Case 1: Title + Series + Book number (after colon)
    match = re.match(r"^(.*?):\s*(.*?),\s*Book\s*(\d+)$", name)
    if match:
        title, series, book_number = match.groups()
        return title.strip(), None, series.strip(), book_number.strip()
    
    # Case 2: Title with #number
    match = re.match(r"^(.*?)\s*#(\d+(?:\s*&\s*\d+)*)$", name)
    if match:
        title, book_number = match.groups()
        return title.strip(), None, None, book_number.strip()
    
    # Case 3: Title + Subtitle (colon but no Book)
    if ":" in name:
        parts = name.split(":", 1)
        return parts[0].strip(), parts[1].strip(), None, None
    
    # Case 4: Just a plain title
    return title.strip(), None, None, None

for name in df["name"] :
    title,subtitle,series,book_number = split_name(name)
    print(f"Title: {title} , subtitle: {subtitle},Series: {series}, Book#: {book_number}")
```

```python
#How can you parse time into consistent numeric fields like duration_minutes (or duration_hours + duration_minutes)?
import re
def parse_duration(x):
    if pd.isna(x):   #check null values
        return none #
    hours_match = re.search(r"(\d+)\s*hr",str(x).lower())
    mins_match = re.search(r"(\d+)\s*min",str(x).lower())
    hours = int(hours_match.group(1)) if hours_match else 0
    mins = int(mins_match.group(1)) if mins_match else 0
    return hours*60 +mins
df["duration_minutes"] = df["time"].apply(parse_duration)# Step 3: Apply to your column (replace "time" with your column name

# Optional: Split into hours + leftover minutes
df["duration_hours"] = df["duration_minutes"] // 60
df["duration_only_minutes"] = df["duration_minutes"] % 60  
print(df["duration_minutes"])
print(df["duration_only_minutes"])
```

```python
#How can you parse stars to extract a numeric rating (e.g., 4.5) and the number of ratings (e.g., 34)? What about when spacing or words are missing?
def parse_stars(x):
    if pd.isna(x): 
        return None, None
    
    text = str(x).lower().replace(",", "")
    
    # find rating (decimal or integer)
    rating_match = re.search(r"(\d+(?:\.\d+)?)", text)
    rating = float(rating_match.group(1)) if rating_match else None
    
    # find number of ratings (number that comes after rating OR in parentheses)
    # second number in string
    numbers = re.findall(r"(\d+(?:\.\d+)?)", text)
    count = int(numbers[1]) if len(numbers) > 1 else None
    
    return rating, count

# Apply to dataframe
df[["rating", "num_ratings"]] = df["stars"].apply(lambda x: pd.Series(parse_stars(x)))

print(df)
```

```python
#Are there near-duplicates (same title + author but slightly different punctuation or whitespace)? How will you detect and consolidate them?
import re
import pandas as pd
df= pd.read_csv(r"C:\Users\amnaj\OneDrive\Desktop\audible_uncleaned (1).csv")
def normalize_text(s):
    if pd.isna(s): return None
    s = s.lower()
    s = re.sub(r"\s+", " ", s.strip())         # collapse spaces
    s = s.replace("’", "'").replace("‘", "'")  # normalize quotes
    s = re.sub(r"[^\w\s']", "", s)             # drop punctuation (except apostrophe)
    return s
df["narrator_clean"] = df["narrator"].apply(normalize_text)
df["author_clean"] = df["author"].apply(normalize_text)
print(df["author_clean"])
print(df["narrator_clean"])
```

```python
#to detect format first then further
import pandas as pd
df= pd.read_csv(r"C:\Users\amnaj\OneDrive\Desktop\audible_uncleaned (1).csv")
def detect_and_parse(date_str):

    if pd.isna(date_str):
        return pd.NaT

    try:
        parts = date_str.split("-")
        if len(parts) != 3:
            return pd.NaT

        first, second, year = map(int, parts)

        if first > 12:  
            return pd.to_datetime(date_str, format="%d-%m-%y", errors="coerce")
        if second > 12: 
            return pd.to_datetime(date_str, format="%m-%d-%y", errors="coerce")
        dt = pd.to_datetime(date_str, format="%d-%m-%y", errors="coerce")
        if pd.isna(dt):
            dt = pd.to_datetime(date_str, format="%m-%d-%y", errors="coerce")
        return dt

    except Exception:
        return pd.NaT
df["release_date_clean"] = df["releasedate"].apply(detect_and_parse)
print(df[["release_date_clean"]])
```
